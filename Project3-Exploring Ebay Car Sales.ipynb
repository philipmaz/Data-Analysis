# Initial Data Exploration and Cleaning

```python
autos.describe(include='all')
```

* Columns that have mostly one value are candidates need to be dropped: 'seller', 'offertype'
* Columns that need more investigation: 'nrofpictures'
* Examples of numeric data stored as text need to be cleaned: 'price' and 'odometer' columns

```python
autos['nrofpictures'].value_counts()
```

**It looks like the 'nrofphotos' column has 0 for every column. It needs to be removed along with the other two columns 'seller' and 'offertype'**

```python
cols_to_drop=['nrofpictures','seller','offertype']
autos_dropped=autos.drop(cols_to_drop,axis=1)
autos_dropped.head()
```

```python
autos_dropped['price']=autos_dropped['price'].str.replace('$','').str.replace(',','').astype(int)
```

```python
autos_dropped['odometer']=autos_dropped['odometer'].str.replace('km','').str.replace(',','').astype(int)
```

```python
autos_dropped.rename({'odometer':'odometer_km'},axis=1,inplace=True)
autos_dropped.rename({'price':'price_$'},axis=1,inplace=True)
autos_dropped.head()
```

**further exploring of 'price_$' and 'odometer' columns:**

```python
autos_dropped['price_$'].describe()
```

```python
autos_dropped['price_$'].unique().shape
```

```python
autos_dropped['price_$'].value_counts().sort_index(ascending=True).head()
```

```python
#autos_dropped=autos_dropped[autos_dropped['price_$'].between(1,351000)]  more simple way to put that
autos_dropped=autos_dropped[(autos_dropped['price_$']>1) & (autos_dropped['price_$']<351000)]
autos_dropped['price_$'].describe()
```

```python
autos_dropped['odometer_km'].value_counts()
```

**We can see that the values in this field are rounded, which might indicate that sellers had to choose from pre-set options for this field. Additionally, there are more high mileage than low mileage vehicles.**

# Dealing with Incorrect Registration Year Data

```python
autos_dropped["registration_year"].describe()
```

**The year that the car was first registered will likely indicate the age of the car. Looking at this column, we note some odd values. The minimum value is 1000, long before cars were invented and the maximum is 9999, many years into the future**

Because a car can't be first registered before the listing was seen, any vehicle with a registration year above 2016 is definitely inaccurate. Determining the earliest valid year is more difficult. Realistically, it could be somewhere in the first few decades of the 1900s.

```python
autos_dropped['registration_year'].value_counts().sort_index()
```

**Given that number of registrations earlier than 1900 and later that 2019 is less than 4% of our data, we will remove these rows.**

```python
(~autos_dropped['registration_year'].between(1910,2016)).sum()/autos_dropped.shape[0]
```

```python
autos_dropped=autos_dropped[autos_dropped['registration_year'].between(1910,2016)]
```

```python
autos_dropped['registration_year'].value_counts(normalize=True).head(10)
```

**It looks like most of the vehicles were registered in the past 20 years**

```python
autos_dropped['registration_year'].describe()
```

# Exploring car brands

```python
autos_dropped['brand'].value_counts(normalize=True)
```


**German manufacturers represent four out of the top five brands, almost 50% of the overall listings. Volkswagen is by far the most popular brand, with approximately double the cars for sale of the next two brands combined.**

**There are lots of brands that don't have a significant percentage of listings, so we will limit our analysis to top 20 brands.**

```python
top20brands=autos_dropped['brand'].value_counts(normalize=True).head(20)
top20brands
```

# Mean price by brand calculation

```python
brands=top20brands.index
mean_price_by_brand={}

for b in brands:
    selected_brands=autos_dropped[autos_dropped['brand']==b]
    brand_mean=selected_brands['price_$'].mean()
    mean_price_by_brand[b]=int(brand_mean)

mean_price_by_brand

```

**Of the top 5 brands, there is a distinct price gap:**

**Audi, BMW and Mercedes Benz are more expensive.
Ford and Opel are less expensive.
Volkswagen is in between - this may explain its popularity, it may be a 'best of 'both worlds' option.**

# Exploring mileage

```python
mean_milage_by_brand={}

for b in brands:
    selected_brands=autos_dropped[autos_dropped['brand']==b]
    mean_milage=selected_brands['odometer_km'].mean()
    mean_milage_by_brand[b]=mean_milage

mean_milage_by_brand
```

```python
brand_mean_price=pd.Series(mean_price_by_brand)
brand_mean_price
```

```python
brand_mean_mileage=pd.Series(mean_milage_by_brand)
brand_mean_mileage
```

```python
df=pd.DataFrame(brand_mean_price,columns=['mean_price'])
df
```

```python
df['mean_mileage']=brand_mean_mileage
df
```

**The range of car mileages does not vary as much as the prices do by brand, instead all falling within 10% for the top brands.**

**There is a slight trend to the more expensive vehicles having higher mileage, with the less expensive vehicles having lower mileage.**
